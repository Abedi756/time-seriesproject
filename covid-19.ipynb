```python
import pandas as pd
```

```python
df = pd.read_csv("time_series_covid19_confirmed_global.csv")
df
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df = df.drop(columns=['Province/State','Lat','Long'])
df
```

```python
df = df.groupby('Country/Region').sum()
df
# this code is the same: df = df.groupby('Country/Region').agg('sum')
```

```python
# create DateTime as the index column
df =df.T
df
```

```python
# change index to Datetime
df.index = pd.to_datetime(df.index)
df
```

```python
df.isnull().sum()
```

# 
## Resampling Time-Series Dataframe

### Here list of most used frequency:
#### 'nM' ===> n months
#### 'nW' ===> n weeks
#### 'nD' ===> n days
#### 'nH' ===> n hours
#### 'nT' or 'nmin' ===> n mins
#### 'nS' ===> n seconds
#### 'nL' or 'nms' ===> n milisoconds
#### 
#### mix together:
#### 'nD.mH' ===> n days, m hours
# 
# 





## Here a list of most used< aggregation-operations>
### summarize: df.resample(timeinterval).sum()
### Mean: df.resample(timeinterval).mean()
### Max: df.resample(timeinterval).max()
### Min: df.resample(timeinterval).min()
### First value: df.resample(timeinterval).first()
### Last value: df.resample(timeinterval).last()
### Standard deviation: df.resample(timeinterval).std()
# 
# 

```python
df.resample('w').mean()
```

```python
df.resample('w').mean().plot()
```

```python
df.resample('w').mean().plot(marker="v",figsize=(15,5))
```

```python
#10 days
df.resample('10D').mean().plot(marker="v",figsize=(15,5))
```

# 
## Percentage Growth

```python
# let's find weekly percentage growth
df.resample('w').mean().pct_change()
```

```python
df.resample('w').mean().pct_change().plot(marker="v",figsize=(15,5))
```

```python
df.resample('m').mean().pct_change().plot(marker="v",figsize=(15,5))
```

# 
## Identify Trends in Time-Series Dataframe

```python
# let's plot the rolling average o confirmed covid cases,by window size o 10 days
df.rolling('10D').mean().plot(marker="v",figsize=(15,5))
```
